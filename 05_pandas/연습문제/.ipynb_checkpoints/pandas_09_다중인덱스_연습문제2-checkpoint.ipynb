## 데이터 통합 및 변형 연습문제

필요 모듈 임포트

```python
import numpy as np
import pandas as pd
```

#### house 폴더에 있는 세 도시의 가구별 데이터 파일들을 사용하여 도시별 데이터프레임을 생성하시오.

- 나주시

```python
naju = pd.read_csv('../data/house/나주시_주택현황 및 보급률(2004~2020).csv', encoding='CP949', header=[0,1,2])
```

```python
naju
```

```python
naju1 = naju.T
naju1
```

```python
naju1 = naju1.unstack(level=0)
naju1
```

```python
naju1.columns = [col[1] for col in naju1.columns]
```

```python
naju1.columns
naju1.index
```

```python
naju1.index = ['일반가구수(A)(가구)', '주택수(B)(호)','단독주택','아파트(호)','연립주택(호)','다세대주택(호)', '비거주용건물내주택(호)', '다가구주택(호)',
               '주택보급률(B)/(A)*100(%)']
naju1
```

- 서산시

```python
seosan1 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2002~2007).csv', encoding='CP949', header=[0,1,2])
seosan2 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2008~2012).csv', encoding='CP949', header=[0,1,2])
seosan3 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2013~2022).csv', encoding='CP949', header=[0,1,2])
```

```python
seosan1 = seosan1.T.unstack(level=0)
seosan2 = seosan2.T.unstack(level=0)
seosan3 = seosan3.T.unstack(level=0)
```

```python
seosan3.index
```

```python
seosan1.drop(columns=[(0, '행정구역(시군)별(1)')], index=[('행정구역(시군)별(1)', '행정구역(시군)별(1)')], inplace=True)
seosan2.drop(columns=[(0, '행정구역(시군)별(1)')], index=[('행정구역(시군)별(1)', '행정구역(시군)별(1)')], inplace=True)
seosan3.drop(columns=[(15, '행정구역(읍면동)별(1)')], index=[('행정구역(읍면동)별(1)', '행정구역(읍면동)별(1)')], inplace=True)
```

```python
seosan1 = seosan1.loc[[('가구수 (가구)','일반가구'),('주택보급률 (%)','소계'),('주택수 (호)','다가구주택'),('주택수 (호)','다세대주택'),
('주택수 (호)','단독주택'),( '주택수 (호)', '비거주용건물내주택'),('주택수 (호)','소계'),('주택수 (호)','아파트'), ('주택수 (호)','연립주택')]]
```

```python
seosan2 = seosan2.loc[[('일반가구수 (가구)','소계'),('주택보급률 (%)','소계'),('주택수 (호)','다가구주택'),('주택수 (호)','다세대주택'),
('주택수 (호)','단독주택'),('주택수 (호)', '비거주용건물 내 주택'),('주택수 (호)','소계'),('주택수 (호)','아파트'), ('주택수 (호)','연립주택')]]
```

```python
seosan3 = seosan3.loc[[('일반가구수 (가구)','소계'),('주택보급률 (%)','소계'),('주택수 (호)','다세대주택'),
('주택수 (호)','단독주택'),('주택수 (호)','단독주택.1'),(   '주택수 (호)', '비거주용 건물내 주택'),('주택수 (호)','소계'),('주택수 (호)','아파트'), ('주택수 (호)','연립주택')]]
```

```python
seosan1.index = ['일반가구수(A)(가구)', '주택보급률(B)/(A)*100(%)', '다가구주택(호)', '다세대주택(호)',
                 '단독주택', '비거주용건물내주택(호)', '주택수(B)(호)', '아파트(호)' , '연립주택(호)']
seosan2.index = ['일반가구수(A)(가구)', '주택보급률(B)/(A)*100(%)', '다가구주택(호)', '다세대주택(호)',
                 '단독주택', '비거주용건물내주택(호)', '주택수(B)(호)', '아파트(호)' , '연립주택(호)']
```

```python
seosan3.index = ['일반가구수(A)(가구)', '주택보급률(B)/(A)*100(%)', '다세대주택(호)',
                 '단독주택','단독주택1', '비거주용건물내주택(호)', '주택수(B)(호)', '아파트(호)' , '연립주택(호)']
```

```python
seosan4 = pd.concat([seosan1,seosan2,seosan3], join='outer', axis=1)
```
