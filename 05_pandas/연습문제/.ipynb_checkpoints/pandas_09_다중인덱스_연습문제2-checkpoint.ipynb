## 데이터 통합 및 변형 연습문제

필요 모듈 임포트

```python
import numpy as np
import pandas as pd
```

#### house 폴더에 있는 세 도시의 가구별 데이터 파일들을 사용하여 도시별 데이터프레임을 생성하시오.

- 나주시

```python
naju = pd.read_csv('../data/house/나주시_주택현황 및 보급률(2004~2020).csv', encoding='CP949', header=[0,1,2])
```

```python
naju
```

```python
naju.columns= [col.split('.')[0] for col in naju.columns]
```

```python
naju.drop(columns='행정구역(시군)별(1)', inplace =True)
```

```python
naju1 = naju.T.reset_index().set_index(['index',0,1])
```

```python
naju1.unstack(level='index')
```

- 서산시

```python
seosan1 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2002~2007).csv', encoding='CP949')
seosan2 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2008~2012).csv', encoding='CP949')
seosan3 = pd.read_csv('../data/house/서산시_주택현황 및 보급률(2013~2022).csv', encoding='CP949')
```

```python
seosan1.drop(columns='행정구역(시군)별(1)', inplace =True)
seosan2.drop(columns='행정구역(시군)별(1)', inplace =True)
seosan3.drop(columns='행정구역(읍면동)별(1)', inplace =True)
```

```python
seosan1.columns= [col.split('.')[0] for col in seosan1.columns]
seosan2.columns= [col.split('.')[0] for col in seosan2.columns]
seosan3.columns= [col.split('.')[0] for col in seosan3.columns]
```

```python
seosan1.join(seosan2, how='outer')
seosan4 = seosan1.join(seosan3, how='outer')
```

```python
seosan4
```

```python
seosan4.T.reset_index().set_index(['index',0,1])
```

```python
seosan4.T.unstack(level=[0,1])
```
