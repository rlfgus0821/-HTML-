### 연습문제1. 
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과 후반기(7월-12월)실적을 나타내는 데이터 프레임을 작성한 후 결합한다.


- 실적 정보는 "매출","비용","이익"으로 이루어진다.

- 이익=매출-비용 : 가공필드로 생성
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

#### 1) month 변수 생성
- 리스트 컴프리헨션을 사용

```python
month = [f'{m}월' for m in range(1, 13)]
month
```

#### 2) 상반기(1~6월) 매출과 비용 data를 정수 난수를 이용하여 생성

- 1~6월의 매출과 비용은 딕셔너리로 구성하되, 키값은 month변수를 활용
- 매월의 매출과 비용은 numpy의 1~100사이의 정수 난수로 생성

```python
# 매출, 비용 data random  생성
data={month[0]:np.random.randint(100,size=2),
     month[1]: np.random.randint(100,size=2),
      month[2]:np.random.randint(100,size=2),
     month[3]: np.random.randint(100,size=2),
      month[4]:np.random.randint(100,size=2),
     month[5]: np.random.randint(100,size=2)}
data
```

#### 3) 상반기(1~6월)의 매출과 비용 데이터프레임 생성

- 2)번에서 생성한 딕셔너리 데이터로 데이터프레임 생성

```python
# 데이터프레임 생성
data={month[0]:np.random.randint(100,size=2),
     month[1]: np.random.randint(100,size=2),
      month[2]:np.random.randint(100,size=2),
     month[3]: np.random.randint(100,size=2),
      month[4]:np.random.randint(100,size=2),
     month[5]: np.random.randint(100,size=2)}
      
df = pd.DataFrame(data, index=['매출','비용'])
df
```

#### 4) '이익' 행 추가
- '이익' 행이 추가된 데이터 프레임 작성
- 이익 = 매출 - 비용 

```python
# '이익' 행 생성
df.loc['이익'] = df.loc['매출'] - df.loc['비용']
df
```

#### 5) 하반기(7~12월) 매출과 비용 data를 정수 난수를 이용하여 생성

- 7~!2월의 매출과 비용은 딕셔너리로 구성하되, 키값은 month변수를 활용
- 매월의 매출과 비용은 numpy의 1~100사이의 정수 난수로 생성

```python
# 매출, 비용 data random  생성
data={month[6]:np.random.randint(100,size=2),
     month[7]: np.random.randint(100,size=2),
      month[8]:np.random.randint(100,size=2),
     month[9]: np.random.randint(100,size=2),
      month[10]:np.random.randint(100,size=2),
     month[11]: np.random.randint(100,size=2)}
data
```

#### 6) 하반기(7~12월) 실적 데이터프레임 생성
- 매출, 비용으로 구성

```python
# 데이터프레임 생성
df1 = pd.DataFrame(data, index=['매출','비용'])
df1
```

#### 7) 하반기 실적에 '이익' 행 추가
- '이익' 행이 추가된 데이터 프레임 작성
- 이익 = 매출 - 비용 

```python
# 가공 행 이익 생성
df1.loc['이익'] = df1.loc['매출'] - df1.loc['비용']
df1
```

#### 8) 상반기 실적 데이터프레임과 하반기 실적 데이터프레임 연결

```python
df2 = df.join(df1)
df2
```

#### 9) '총실적' 열 추가

```python
df2['총실적'] = df2.sum(axis=1)
df2
```

### 연습문제 3
- 아래 csv 파일을 다운 받아 데이터프레임을 생성하고 인덱싱을 통해 일부를 추출 해 문제에 사용할 데이터 프레임을 추출한다
- 추출된 데이터 프레임을 사용해 제시된 문제를 해결한다

Dataurl = ‘https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv’

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df
```

```python
df1 = df.iloc[:5,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

#### 1) df1과 df2 데이터를 하나의 데이터 프레임으로 합치기

```python
pd.concat([df1,df2])
```

#### 2) df1과 df2를 결합하되 중복행은 한번만 표시되게 결합하시오
- 중복열은 허용 함

```python
pd.concat([df1,df2],axis=1,join='outer')
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

#### 3) df3과 df4 데이터를 하나의 데이터 프레임으로 결합하되 둘다 포함하고 있는 년도에 대해서만 작업을 진행하세요

```python
pd.concat([df3,df4], join='inner')
```

#### 4) df3과 df4 데이터를 하나의 데이터 프레임으로 합치시오. 
- 모든 컬럼을 포함하도록 두 데이터프레임을 합치고
- 결측치는 0으로 대체하되, 모든 데이터값을 정수형으로 변경하시오

```python
pd.concat([df3,df4],axis=1).fillna(0).astype(int)
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

#### 5) df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합치시오 
- Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하세요

```python
df5.merge(df6, on='Algeria', how='inner')
```

- 두 데이터 모두 포함하는 데이터만 출력하되, 두 데이터프레임의 index를 사용하시오.

```python
df5.merge(df6, how='inner', left_index=True, right_index=True)
```

- Algeria컬럼을 key로 하고 합집합으로 합치세요

```python
df5.merge(df6, on='Algeria', how='outer')
```
