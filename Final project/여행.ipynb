```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 범주형 데이터를 숫자로 변환
label_encoders = {}
for column in ['GENDER', 'VISIT_AREA_NM']:
    label = LabelEncoder()
    df_filter[column] = label.fit_transform(df_filter[column])
    label_encoders[column] = label

# 특징과 타겟 변수 분리
x = df_filter.drop('DGSTFN', axis=1)[:10000]
y = df_filter['DGSTFN'][:10000]

# 학습용 데이터와 테스트용 데이터 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 0)

# 모델 리스트
models = {
    'Random Forest': RandomForestClassifier(n_estimators=10, random_state= 0),
    'Logistic Regression': LogisticRegression(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=10, random_state= 0)
}

# 모델 학습 및 평가
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} 모델 정확도: {accuracy:.4f}')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd


x = df_filter.drop('DGSTFN', axis=1)[:10000]
y = df_filter['DGSTFN'][:10000]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# 수치형 특징과 범주형 특징 분리
numeric_features = x_train.select_dtypes(include=['int32', 'float64']).columns.tolist()
categorical_features = ['GENDER', 'VISIT_AREA_NM']

# 전처리 파이프라인
# 수치형 -> 표준화
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
# 범주형 -> 원핫인코딩
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# ColumnTransformer -> 여러 종류의 전처리 작업을 서로 다른 열에 적용할 수 있도록 도와줌
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# 모델 리스트
models = {
    'Random Forest': RandomForestClassifier(random_state=0),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(random_state=0)
}

# 하이퍼파라미터 튜닝 및 모델 학습
best_models = {}
for name, model in models.items():
    pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model)])
    
    if name == 'Random Forest':
        param_grid = {
            'classifier__n_estimators': [50, 100, 200],
            'classifier__max_depth': [None, 10, 20]
        }
    elif name == 'Logistic Regression':
        param_grid = {
            'classifier__C': [0.01, 0.1, 1, 10]
        }
    elif name == 'K-Nearest Neighbors':
        param_grid = {
            'classifier__n_neighbors': [3, 5, 7],
            'classifier__weights': ['uniform', 'distance']
        }
    elif name == 'Gradient Boosting':
        param_grid = {
            'classifier__n_estimators': [50, 100, 200],
            'classifier__learning_rate': [0.01, 0.1, 0.2]
        }
    
    grid_search = GridSearchCV(pipeline, param_grid, cv=3, n_jobs=-1)
    grid_search.fit(x_train, y_train)
    best_models[name] = grid_search.best_estimator_
    y_pred = grid_search.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} 모델 정확도: {accuracy:.4f}')

# 최종 모델 선택 및 평가
for name, model in best_models.items():
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{name} 최적화된 모델 정확도: {accuracy:.4f}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score

# 예제 데이터
area_names = pd.DataFrame({'VISIT_AREA_NM': ['Area1', 'Area2', 'Area3']})

# 피처와 타겟 정의
# df_filter는 학습에 사용한 DataFrame으로 가정
x = df_filter.drop('DGSTFN', axis=1)
y = df_filter['DGSTFN']

# 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# 수치형 특징과 범주형 특징 분리
numeric_features = x_train.select_dtypes(include=['int32', 'float64']).columns.tolist()
categorical_features = ['GENDER', 'VISIT_AREA_NM']

# 전처리 파이프라인 구성
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# 랜덤 포레스트 모델 정의 및 하이퍼파라미터 튜닝
model = RandomForestClassifier(random_state=0)
param_grid = {
    'classifier__n_estimators': [50, 100, 200],
    'classifier__max_depth': [None, 10, 20]
}
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model)])
grid_search = GridSearchCV(pipeline, param_grid, cv=3, n_jobs=-1)

# 모델 학습
grid_search.fit(x_train, y_train)

# 최적 모델 선택 및 평가
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'랜덤 포레스트 최적화된 모델 정확도: {accuracy:.4f}')

# 새로운 입력 데이터 예측
traveler = {'GENDER': 1.0,
  'AGE_GRP': 30.0,
  'TRAVEL_STYL_1': 2.0,
  'TRAVEL_STYL_2': 5.0,
  'TRAVEL_STYL_3': 3.0,
  'TRAVEL_STYL_4': 7.0,
  'TRAVEL_STYL_5': 2.0,
  'TRAVEL_STYL_6': 2.0,
  'TRAVEL_STYL_7': 7.0,
  'TRAVEL_STYL_8': 6.0,
  'TRAVEL_MOTIVE_1': 1.0,
  'TRAVEL_COMPANIONS_NUM': 1.0,
  'TRAVEL_MISSION_INT': 6.0,
  'VISIT_AREA_NM': 'Area1'  # 초기 값은 임의의 값으로 설정
}

results = pd.DataFrame([], columns=['AREA', 'SCORE'])

for area in area_names['VISIT_AREA_NM']:
    traveler['VISIT_AREA_NM'] = area
    input_data = pd.DataFrame([traveler])
    
    # 전처리 및 예측
    input_data_transformed = best_model.named_steps['preprocessor'].transform(input_data)
    score = best_model.named_steps['classifier'].predict(input_data_transformed)
    
    results = pd.concat([results, pd.DataFrame([[area, score[0]]], columns=['AREA', 'SCORE'])])

# 결과 정렬 및 상위 10개 출력
results = results.sort_values('SCORE', ascending=False).head(10)
print(results)
```
