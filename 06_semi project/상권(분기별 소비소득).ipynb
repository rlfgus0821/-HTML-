```python
# 그래프에서 한글폰트 지원
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
pay=pd.read_csv('data/서울시 상권분석서비스(소득소비-행정동).csv', encoding='cp949')
```

```python
pd.DataFrame(pay).isnull().sum()
```

```python
pd.DataFrame(pay)
```

```python
# 2022년도 내용추출
pay.set_index('기준_년분기_코드',inplace=True)
pay1 = pay.loc[[20221,20222,20223,20224]]
pay1.info()
```

```python
pay1.reset_index(drop=False,inplace=True)
pay1
```

```python
# 행정동과 분기년월 인덱스로 변경
indexyear =list(pay1['기준_년분기_코드'])
len(indexyear)
indexname =list(pay1['행정동_코드_명'])
len(indexname)
pay1['기준_년분기_코드']

pay2 = pay1.set_index([indexname, indexyear])
pay2.drop(['행정동_코드_명','기준_년분기_코드'],axis=1,inplace=True)
pay2.head(3)
```

```python
# 중복값이 없다.
pay1['지출_총금액'].value_counts()
```

```python
# 지출_총금액 TOP10
pay1['지출_총금액'].nlargest(10)
```

```python
# 이상한 이름 변경
pay2.rename(index={'종로1?2?3?4가동':'종로1.2.3.4가동'},inplace=True)
```

```python
# columns별 TOP10뽑기
col=[]
columns = ['월_평균_소득_금액', '지출_총금액', '식료품_지출_총금액',
       '의류_신발_지출_총금액', '생활용품_지출_총금액', '의료비_지출_총금액', '교통_지출_총금액', '교육_지출_총금액',
       '유흥_지출_총금액', '여가_문화_지출_총금액', '기타_지출_총금액', '음식_지출_총금액']
for i in columns:
    col.append(pay2[i].nlargest(10))
col
```

```python
# 월_평균_소득_금액 TOP10
col[0].plot.barh(figsize=(4,4), title='월_평균_소득_금액 TOP10')
plt.show()
```
