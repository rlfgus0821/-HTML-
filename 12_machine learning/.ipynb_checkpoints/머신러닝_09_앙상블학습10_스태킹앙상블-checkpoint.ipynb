# 스태킹 앙상블

### 스태킹(Stacking)

- 여러 알고리즘을 결합해 예측 결과를 도출
- 개별 알고리즘으로 예측한 데이터를 기반으로 다시 예측 수행
- 스태킹 모델의 구성 요소
    - 개별적인 기반 모델(base leaner)
    - 최종 메타 모델(meta leaner)
- 스태킹 모델의 핵심 : 여러 개별 모델의 예측 데이터를 각각 스태킹 형태로 결합해 최종 메터 모델의 학습용 피처 데이터 세트와 테스트용 피처 데이터 세트를 만드는 것

### 기본 스태킹 모델

#### 모듈 임포트

#### 예제. 위스콘신 유방암 데이터세트

#### 개별 모델 생성/학습/예측

#### 최종모델 학습을 위한 데이터 스태킹/학습

### CV 세트 기반의 스태킹

- 과적합을 개선하기 위해 최종 메타 모델을 위한 데이터 세트를 만들 때 교차 검증 기반으로 예측된 결과 데이터세트를 이용
- 학습용 스태킹 데이터 생성/테스트용 스태킹 데이터 생성하여 최종 메타 모델이 학습과 예측을 수행
