# 카이제곱 검정

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

%precision 4
```

###  카이제곱 분포

```python
x = np.linspace(0,20,50)
for d in [1,2,3,8,10,20]:
    rv = stats.chi2(df=d)
    y = rv.pdf(x)
    plt.plot(x,y, label= f'df={d}')
plt.legend()
plt.show()
```

## 적합도 검정

- 가설
    - H0: 어떤 분포를 따른다

예제1. 완두콩 교배결과가 멘델의 유전법칙 (9:3:3:1)을 따르는가?

- H0: 멘델의 유전법칙을 따른다.(9:3:3:1)

```python
# 관측 빈도
obs = [322,108,98,32]

# 기대 빈도
mendel = [9,3,3,1]
ratio = mendel / np.sum(mendel)
exp = ratio * np.sum(obs)
exp
```

```python
# 적합도 검정
stat, p = stats.chisquare(obs, exp)
result = 'Accept H0' if p >0.05 else 'Reject H0'
print(f'적합도 검정통계량: {stat:.4f}, p_value: {p:.4f}, 적합도 검정결과: {result}')
```

예제2. 주사위 눈의 분포는 균일분포를 따르는지 적합도 검정을 수행하시오.

- H0: 각 눈이 나올 확률은 동일하다.

```python
# 관측값
obs = [16,18,16,14,12,12]

# 기대값
dice = [np.sum(obs)/6] * 6
dice
```

```python
# 적합도 검정
stat, p = stats.chisquare(obs, dice)
result = 'Accept H0' if p >0.05 else 'Reject H0'
print(f'적합도 검정통계량: {stat:.4f}, p_value: {p:.4f}, 적합도 검정결과: {result}')
```
